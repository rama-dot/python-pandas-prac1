```python
# Create a DataFrame with the data above

dic = {'sales' : [100,130,119,92,35], 'customer_account' : ['B100','J101','X102','P103','R104'], 'city' : ['BOS','LA','NYC','SF','CHI']
      , 'profit' : [200,300,410,530,490]}
df = pd.DataFrame(dic)
df
```

```python
# Sort the DataFrame by profit in descending order

df.sort_values('profit',ascending = False)
```
